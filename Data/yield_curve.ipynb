```python
treasury_data = Path("Data/Treasury_Yield_Curve_Rates.csv")
treasury_df = pd.read_csv(treasury_data)
treasury_df.rename(columns={'Date':'date'}, inplace=True)
treasury_df.head()
```

```python
treasury_df['date'] = pd.to_datetime(treasury_df['date'])
treasury_df = treasury_df.set_index('date')
treasury_df.head()
```

```python
treasury_df = treasury_df.dropna(axis='rows')
treasury_df.head()
```

```python
treasury_df.plot()
```

```python
treasury_df_transposed = treasury_df.T
treasury_df_transposed.head()
```
