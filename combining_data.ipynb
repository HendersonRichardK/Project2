```python
# Import libraries and dependencies
import pandas as pd
from pathlib import Path
%matplotlib inline
import os
import pandas as pd
import matplotlib.pyplot as plt
```

# DJIA

```python
djia_data = Path("Data/1_data_cleanup/djia.csv")
djia_df = pd.read_csv(djia_data)
djia_df.rename(columns={'Date':'date'}, inplace=True)
djia_df['date'] = pd.to_datetime(djia_df['date'])
djia_df = djia_df.set_index('date')
djia_df = djia_df.rename(columns={'Price': 'DJIA', 'Change %' : 'DIJA_pct_change', 'Vol.' : 'DIJA_vol_in_millions', 'Low' : 'DIJA_Low', 'High' : 'DIJA_high', 'Open' : 'DIJA_open'}).sort_values(by='date')
djia_df = djia_df.loc["2016-01-01":"2020-06-30"]
djia_df['DJIA'] = djia_df['DJIA'].str.replace(',', '').astype(float)
djia_df['DIJA_pct_change'] = djia_df['DIJA_pct_change'].str.replace('%', '').astype(float)
djia_df['DIJA_vol_in_millions'] = djia_df['DIJA_vol_in_millions'].str.replace('M', '').astype(float)
djia_df['DIJA_Low'] = djia_df['DIJA_Low'].str.replace(',', '').astype(float)
djia_df['DIJA_high'] = djia_df['DIJA_high'].str.replace(',', '').astype(float)
djia_df['DIJA_open'] = djia_df['DIJA_open'].str.replace(',', '').astype(float)
djia_df.head()
```
